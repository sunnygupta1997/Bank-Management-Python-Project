```python
class Bank:
  def __init__(self,name,age,aadhar):
    self.Name=name
    self.Age=age
    self.Aadhar=str(aadhar)
    self.Balance=0

  # Deposit
  def deposit(self,amount):
    print("------------------------------")

    if amount <= 0:
      print("❌ Invalid amount! Negative deposit not allowed")
      print("------------------------------")
      return

    print("Adding",amount,"to balance.")
    self.Balance = self.Balance + amount
    print("Current Balance:",self.Balance)
    print("------------------------------")

  # Withdraw
  def withdraw(self,amount):
    print("------------------------------")

    if amount <= 0:
      print("❌ Invalid amount!")

    elif amount > self.Balance:
      print("Insufficient Balance")

    else:
      print("Getting",amount,"from balance")
      self.Balance = self.Balance - amount
      print("Current Balance:",self.Balance)

    print("------------------------------")

  # Check Balance
  def check_balance(self):
    print("------------------------------")
    print("Available Balance:",self.Balance)
    print("------------------------------")

  # Interest Calculation
  def add_interest(self,rate,years):
    print("------------------------------")
    interest = (self.Balance * rate * years) / 100
    self.Balance = self.Balance + interest
    print("Interest added:",interest)
    print("New Balance:",self.Balance)
    print("------------------------------")


# Object
Nishant = Bank("Nishant",25,526474839237)

# Testing
Nishant.deposit(100000)
Nishant.deposit(-500)   # negative deposit
Nishant.check_balance()
Nishant.add_interest(5,2)   # 5% for 2 years
Nishant.withdraw(20000)
Nishant.check_balance()
```
